### Load MNIST

```python
from keras.datasets import mnist
```

```python
(trainX, trainy), (testX, testy) = mnist.load_data()
```

```python
trL = trainX.astype('float').reshape((60000,784))
```

### Create default algo and plotter functions

```python
def plotter(results,labels,subtitle,axis,legend =True):
    lcount = len(set(labels))
    s = sns.scatterplot(results[:,0],results[:,1],
                    hue = labels,
                    palette=sns.color_palette("hls",lcount),
                    ax = axis)
    if legend==True:
        s.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
    else:
        s.legend().set_visible(False)
    s.set_title(subtitle)
    
def pca(data):
    return PCA(n_components = 2).fit_transform(data)

def tsne(data,perplexity = 40):
    return TSNE(n_components = 2, verbose = 1, perplexity = perplexity, n_iter = 2000).fit_transform(data)

def isomap(data):
    return Isomap(n_components= 2).fit_transform(data)

def lle(data):
    return LocallyLinearEmbedding(n_components =2).fit_transform(data)

def sammon_wr(data):
    return sammon(data,2,display=0)[0]
```

```python

def compare(data,labels,data_label,algos):
    print(f"Comparing {list(algos.keys())} algos, with {data_label} data with {len(data)} obs at {datetime.datetime.now()}")
    data = PCA(n_components = 30).fit_transform(data)
    
    fig, ax = plt.subplots(3, 2,figsize=(15,15))
    i = 0
    j = 0
    for algo_label,algo in algos.items():
        start = datetime.datetime.now()
        results = algo(data)
        end = datetime.datetime.now()
        time = end-start
        #print(round(time.seconds /60,5))
        subtitle = algo_label + f" ({round((time.seconds+time.microseconds*1e-6)/60,5)} minutes)"
        plotter(results,labels,subtitle, ax[i][j])
        j+=1
        if j ==2:
            j=0
            i+=1
    plt.savefig("figs/MNIST_algos2.png")
    plt.show()
        
algos = {"PCA":pca,
         "ISOMAP":isomap,
         "LLE":lle,
         "Sammon":sammon_wr,
        "t-SNE":tsne}

compare(trL[:5000,:],trainy[:5000],"MNIST",algos)
```

# Tuning t-sne

```python
%%capture
import random
random.seed(42)
N = 5000
fig, ax = plt.subplots(3, 2,figsize=(15,15))
i = 0
j = 0
rndperm = np.random.permutation(trL.shape[0])[:N]
raw = trL[rndperm,:]

data = PCA(n_components = 30).fit_transform(raw)
for perp in [5,15,25,35,45,55]:
    results= tsne(data,perplexity = perp)
    plotter(results,
           trainy[rndperm],
           f"perplexity = {perp}",
          ax[i][j])
    j+=1
    if j ==2:
        j=0
        i+=1
plt.savefig("figs/tsne_MNIST_tuning.png")
plt.show()
```

#### Crowding Problem Toy Example

```python
corner = np.array([[1,0],[0,0],[0,1]])
flat = np.array([[1,0],[0,0],[-1,0]])
label = ["3","2","1"]
fig, ax = plt.subplots(1, 2,figsize=(10,5))
i = 0
for d in [corner, flat]:
    ax[i].scatter(d[:,0],d[:,1],s = 100)
    ax[i].plot(d[:,0],d[:,1])
    for j, txt in enumerate(label):
        ax[i].annotate(txt, (d[j,0]+0.1,d[j,1]+0.1))
    
    ax[i].set_xlim(-1.5,1.5)
    ax[i].set_ylim(-1.5,1.5)
    i+=1
plt.savefig("figs/crowding.png")
plt.show()
```

### Create Weakness Plots

```python
N = 100
x_big = np.random.multivariate_normal(np.array([0,0]),np.identity(2)*1,size = N)
x_small = np.random.multivariate_normal(np.array([4,0]),np.identity(2)*0.1,size = N)
labels = np.array([0]*N + [1]*N)
X_var = np.concatenate((x_big,x_small),axis = 0)

print(X_var.shape)
```

```python
%%capture
def weakness_plt(X,labels,name):
    fig, ax = plt.subplots(2, 3,figsize=(10,6))
    fig.tight_layout()
    i = 0
    j = 1
    for perp in [5,15,25,35,45]:
        results= tsne(X,perplexity = perp)
        plotter(results,
               labels,
               f"perplexity = {perp}",
              ax[i][j],legend = False)
        j+=1
        if j ==3:
            j=0
            i+=1
    plotter(X,labels,f"Original Data",ax[0][0],legend = False)
    #plt.get_legend().remove()
    plt.savefig(f"figs/weakness_{name}.png")
    plt.show()
    
weakness_plt(X_var,labels,"cluster_density")
```

```python
Image(filename='figs/weakness_cluster_density.png')
```

```python
x_c1 = np.random.multivariate_normal(np.array([4,4]),np.identity(2)*0.1,size = N)
x_c2 = np.random.multivariate_normal(np.array([3,3]),np.identity(2)*0.1,size = N)
x_c3 = np.random.multivariate_normal(np.array([-3,-3]),np.identity(2)*0.1,size = N)
labels = np.array([0]*N + [1]*N + [2]*N)
X_dist = np.concatenate((x_c1,x_c2,x_c3),axis = 0)
```

```python
%%capture
weakness_plt(X_dist,labels,"cluster_distance")
```
